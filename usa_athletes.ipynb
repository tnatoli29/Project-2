```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data set athletes.csv is a data set of all athletes from the Rio de Janiero 2016 Olympics
athletes_file = "Resources/athletes.csv"
athletes_2016 = pd.read_csv(athletes_file)
athletes_2016
```

```python
# Add Year column into the athletes dataset to show '2016'
# Add City column to show where the Olympic city
athletes_2016['year'] = 2016
athletes_2016['city'] = 'Rio de Janeiro'

athletes_2016
```

```python
# Select columns from the data set athletes.csv 
all_athletes_2016 = athletes_2016[['city', 'year', 'name','nationality', 'sex', 'gold', 'silver', 'bronze']]
all_athletes_2016
```

```python
# Filter only USA athletes from the 2016 dataset
usa_athletes_2016 = all_athletes_2016[all_athletes_2016.nationality == "USA"]
usa_athletes_2016
```

```python
# Final USA Athletes for 2016 dataframe
final_usa_athletes_2016 = usa_athletes_2016[['city', 'year', 'name','nationality', 'sex']]
final_usa_athletes_2016
```

```python
# BREAK FOR NEXT DATASET - Summer Olympics for 1976 to 2008
```

```python
# Summer Olympic dataset from years 1976 to 2008
summer_file = "Resources/Summer-Olympic-medals-1976-to-2008.csv"
summer_athletes = pd.read_csv(summer_file)
summer_athletes
```

```python
# Select columns from the data set Summer 1976 to 2008.csv 
summer_athletes = summer_athletes[['City', 'Year', 'Athlete','Gender', 'Country_Code', 'Medal']]
summer_athletes
```

```python
# Filter only USA athletes Summer 1976 to 2008 dataset
usa_summer = summer_athletes[summer_athletes['Country_Code'] == "USA"]
usa_summer
```

```python
# Using .rename(columns={}) in order to rename columns
summer_usa_only = usa_summer.rename(columns={"City":"city", "Year":"year", "Gender":"sex", 
                                       "Country_Code":"nationality", "Medal":"medal"})
summer_usa_only.head()
```

```python
#Change variables values  
summer_usa_only['sex'] = summer_usa_only['sex'].replace(["Men"],'male')
summer_usa_only['sex'] = summer_usa_only['sex'].replace(["Women"],'female')
summer_usa_only
```

```python
#Clean data a bit more
new = summer_usa_only["Athlete"].str.split(",", n=1, expand=True)
summer_usa_only["Last_name"] = new[0]
summer_usa_only["First_name"] = new[1]
summer_usa_only.drop(columns = ["Athlete"], inplace = True)
summer_usa_only["Last_name"] = summer_usa_only["Last_name"].str.capitalize()
summer_usa_only
```

```python
#Grab first names and last names and combine them together again 
summer_usa_only["name"] = summer_usa_only["First_name"] +" " + summer_usa_only["Last_name"]
del summer_usa_only['Last_name']
del summer_usa_only['First_name']

summer_usa_only
```

```python
# Final USA Athletes from Summer 1976 to 2008
final_usa_summer = summer_usa_only[['city', 'year', 'name','nationality', 'sex']]
final_usa_summer
```

```python
# BREAK FOR NEXT - MERGE 2 FILES INTO ONE
```

```python
# Merge files together into 1
#merge = pd.concat([final_usa_athletes_2016, final_usa_summer], axis = 0, join; str = "outer")
frames= [final_usa_athletes_2016, final_usa_summer]
usa = pd.concat(frames)
usa.reset_index(inplace=True, drop=True)
usa = usa.astype({'year': 'int'})
usa
```

```python
usa['sex'].value_counts()
```

```python
# Split geneder athletes
female = usa[usa.sex == "female"]
male = usa[usa.sex == 'male']
```

```python
male_count_by_year = male['year'].value_counts()
female_count_by_year = female['year'].value_counts()
```

```python
reset_male = male_count_by_year.reset_index()
reset_male.columns = ['year', 'male']
reset_male
```

```python
reset_female = female_count_by_year.reset_index()
reset_female.columns = ['year', 'female']
reset_female
```

```python
# Merge the two male and female count by year files for plotting

merge_gender_files = pd.merge(reset_male, reset_female, on="year")
gender_by_year = merge_gender_files
gender_by_year
```

```python
gender_by_year["total"] = gender_by_year.male + gender_by_year.female
gender_by_year
```

```python
# Sort by Year in ascending order
gender_by_year.sort_values('year', axis=0, ascending=True, inplace=True, na_position='first')
gender_by_year
```

```python
# Save gender_by_year as a csv file for html plotting
#gender_by_year.to_csv("Resources/totalUSAthletes.csv", index=False)
```

```python
# Create lists for plotting
year = gender_by_year['year'].tolist()
total = gender_by_year['total'].tolist()
male = gender_by_year['male'].tolist()
female = gender_by_year['female'].tolist()
```

```python
# Plot the charts and apply some styling
# Plot based on Summer Olympic files, all athletes
male, = plt.plot(year, male, marker="o", color="blue", label="Male Athletes")
female, = plt.plot(year, female, marker="o", color="red", label="Female Athletes")

plt.title("Olympic USA Athletes from 1976 to 2016")
plt.xlabel("Olympic Year")
plt.ylabel("Total Number of USA Athletes")
plt.legend(handles=[male, female], loc="best")
```
